```python
# import necessary libraries

import numpy as np # for linear algebra
import pandas as pd # for creating dataframes

from statsmodels.formula.api import ols # for Ordinary Least Squares method
```

```python
# import the dataset

survey_data = pd.read_csv('dataSurvey.csv')
```

```python
# preview the dataset

survey_data.head(5)
```

```python
#  drop unnecessary columns (if needed)

survey_data = survey_data.drop(columns='Unnamed: 4')
```

```python
# preview the dataset again

survey_data.head(5)
```

```python
# specify the regression with OLS

model_for_survey = ols(formula='Rating~attendance+avg_test_score+sleep_hours', data=survey_data).fit()
```

```python
# display the parameters of the model 
# add '.round(3)' after 'params' to have them rounded to 3 decimal places

model_for_survey.params
```
