## Estimate the regression

### Non-robust vs Robust

A robust test in econometrics usually means conducting standard hypothesis tests using variance estimators that remain consistent under heteroskedasticity, serial correlation, or clustering. 

Robust test differs from a non-robust test in the command '.fit()'. Here we will be fitting a model with a robust HC3 estimate of the covariance matrix.

To carry out a robust test add ".fit(cov_type='HC3')".

The command ".fit(cov_type='HC3')" implies computing heteroskedasticity-robust standard errors using the HC3 covariance estimator.

Doing a robust or a non-robust test is taken account when evaluating a regression. The steps after it are communicated this adjustment from estimating the regression.

```python
# Non-robust
pancakes_should_be_eaten_model_1 = ols(formula='sleep~totwrk+age+np.square(age)+south+male+smsa', data=pancake_yummy).fit()
```

```python
# Robust
cupcakes_should_be_eaten_model_2 = ols(formula='sleep~totwrk+age+np.square(age)+south+male+smsa', data=pancake_yummy).fit(cov_type='HC3')
```

## Significance of the entire regression

#### Get the F-statistic value

```python
# Use this function to get the F-statistic value for the whole model
# model.fvalue
pancakes_should_be_eaten_model_1.fvalue.round(3)
```

#### Calculate the critical value

```python
# significance_level (α)= 0.05 = 5%
# dfd = degrees_of_freedom = n-k-1 = 706-7-1 (for this specific case)
# n - number of observations
# dfn = k - number of regressors in the model (number of X's)
# f.isf(q=α, dfn=k, dfd=n-k-1)
f.isf(q=0.05,dfn=6,dfd=706-6-1).round(3)
```

#### Compare the F-statistic value with the critical value

If |F_stat| < F_cr => we fail to reject the null hypothesis (H0) => the regression is not statistically significant (at α significance level). The B coefficients can be taken equal to 0.

If |F_stat| > F_cr => we reject the null hypothesis (H0) => the regression is statistically significant (at α significance level). At least one of the B coefficients canbot be taken equal to 0.

#### OR compare the P-value with the significance level

If the p-value > alpha/α/significance level => we fail to reject the H0.

If the p-value < alpha/α/significance level => we to reject the H0.

```python
# Use this function to get the P-value for the whole model
# model.f_pvalue
pancakes_should_be_eaten_model_1.f_pvalue
```

## Significance of several coefficients

With the F-test we can estimate the statistical significance of several coefficents at once, not necessarily the whole model.

#### Get the F-statistic value

```python
# To run the F-test for several regressors use the following function
# model.f_test(X1=X2=0)
# OR
# model.f_test(X1=0, X2=0)
# Where substitute X1, X2, ... with the regressors from the model
# If you have np.square(X) or np.log(X), include them with np. functions
pancakes_should_be_eaten_model_1.f_test('totwrk=age=0')
```

The results of the F-test for several coefficients show
- the F-statistic value for the coefficients selected from the model
- the P-value
- the degrees of freedom

#### Calculate the critical value

While determining statistical signifficance of several coefficients you need to calculate the critical value with the parameters **df_denom** and **df_num** from the F-test results.

```python
f.isf(q=0.05, dfn=2,dfd = 699).round(2)
```
