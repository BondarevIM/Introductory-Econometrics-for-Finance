```python
from scipy.stats import t # import t-distribution and critical values
```

## **Calculate the critical value**

```python
# Option 1
# significance_level (α)= 0.05 = 5%
# degrees_of_freedom = n-k-1 = 706-7-1 (for this specific case)
# n - number of observations
# k - number of regressors
# t.ppf
t.ppf(q=1-0.05/2,df=706-7-1)
```

```python
# Option 2
# t.isf
t.isf(q=0.05/2,df=706-7-1).round(3)
```

#### The critical value is 1.963

#### t_cr (at 5% sign. level) = 1.963

## **Which coefficients are significant?**

### Using t_stat and t_cr (recommended)

Use the **absolute value of the t-statistic** and compare it to the positive critical value. (We consider the case of the Two-sided T-test/Two-Tailed T-test)

If |t_stat| < t_cr => we fail to reject the null hypothesis (H0) => the X is **not correlated** with Y in the population (at α significance level)

If |t_stat| > t_cr => we reject the null hypothesis (H0) => the X is **correlated** with Y in the population (at α sign level)

•	totwrk t_stat = -9.363<br>
•	totwrk t_stat module = |-9.363| = 9.363<br>
•	totwrk t_cr = 1.963<br>
Answer: If **totwrk** |t_stat| > t_cr (|-9.363| > 1.963) => we reject the null hypothesis (H0) => the **totwrk** is correlated with **sleep** in the population (at **5%** significance level).

### Another way:<br>Compare to the p-value


If the p-value > alpha/α/significance level => we fail to reject the null (H0) => the X is **not correlated** with Y in the population (at α sign level)

If the p-value < alpha/α/significance level => we reject the null hypothesis (H0) => the X is **correlated** with Y in the population (at α sign level)

Answer: If **totwrk** P-value < α (0.000 < 0.05) => we reject the null hypothesis (H0) => the **totwrk** is correlated with **sleep** in the population (at **5%** significance level).
