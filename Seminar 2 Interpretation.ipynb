## **Definitions**

Parameter / Coefficient – B

Regressor / Explanatory Variable – X

Dependent Variable - Y

## **Interpretation**

### **Blueprint:**

#### An increase in **X** by 1 unit/ by 1% holding others constant is associated with a **B** units/% increase/descrease in **Y** on average.

**Remember:** <br>
We interpete the X, not log(X) or anything else.<br>
*Example:* "By how much will the profit rise?" in relation to log(X)

### Y = B0 + B1\*X1

**Interpretation:** An increase in **X** by 1 unit holding others constant is associated with a **B** units increase/descrease in **Y** on average.

### log(Y) = B0 + B1\*X1

**Interpretation:** An increase in **X** by 1 unit holding others constant is associated with a **B**\*100% increase/descrease in **Y** on average.

### Y = B0 + B1\*log(X1)

**Interpretation:** An increase in **X** by 1% holding others constant is associated with a **B**\/100% increase/descrease in **Y** on average.

### log(Y) = B0 + B1\*log(X1)

**Interpretation:** An increase in **X** by 1% holding others constant is associated with a **B**% increase/descrease in **Y** on average.
