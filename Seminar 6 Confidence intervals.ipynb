## Calculate the confidence intervals for each coefficient separately

The confidence interval helps us to determine statistical significance of an individual regressor at a certain significance level.

If the confidence interval **includes zero** we fail to reject H0.<br>
If the confidence interval **excludes zero** we reject H0.

Usually H0: β = 0

Confidence intervals are intervals first of all. So, they include two points.<br>
A confidence interval can be interpreted as 'with α% certainty the effect of the X for the Y lies within the interval \[...\]'.

I = (β - Std_err \* t_cr; β + Std_err \* t_cr)<br>
The lower boundary: β-Std_err\*t_cr<br>
The upper boundary: β+Std_err\*t_cr

**Examples for α = 5%:**<br>
A 95% confidence interval that excludes zero ⇔ Reject H0: βj = 0 at  α = 0.05<br>
A 95% confidence interval that includes zero ⇔ Fail to reject H0: βj = 0 at α = 0.05

### Using the built-in method for all regressors at once
