```python
# import necessary libraries

import numpy as np # for linear algebra
import pandas as pd # for creating dataframes

from statsmodels.formula.api import ols # for Ordinary Least Squares method
```

```python
# import the dataset
# you create a dataframe called anyhow you like
# read_csv saves the csv data to a dataframe
# .head(n) function displays top n rows of the dataframe
# .head() displays top 5 rows by default

AppleS_for_SAleeee3 = pd.read_csv('sleep75.csv')
AppleS_for_SAleeee3.head()
```

```python
# display the list of dataset columns names to copy them as explaining variables/regressors/'x1', 'x2', 'x3', ...

AppleS_for_SAleeee3.columns.tolist()
```

```python
# specify the regression with OLS
# you create an object called anyhow you like
# 'formula'specifies the regression formula
# 'data' specifies the data and must correspond to the dataframe
# regressors ('x1', 'x2', 'x3', ...) in 'formula' must correspond to the dataframe columns in 'data'
# the variable 'y' (depending variable) is written before '~' ('~' is the same as '=' in the equation)
# successful execution of the cell will result in blank output (the model will be created and memorized)

Mgi1m0_iz_da_b3SSt = ols(formula='sleep~educ+age+hrwage', data=AppleS_for_SAleeee3).fit()
```

```python
# .summary() displays the complete summary of the model

Mgi1m0_iz_da_b3SSt.summary()
```

```python
# displays the regression's parameters/coefficients before 'x1', 'x2', 'x3', ...
# .round(n) rounds the coefficients to n decimal places
# .round(6) seems to be the limit of possible decimal places

Mgi1m0_iz_da_b3SSt.params.round(3)
```

## **New model for prediction for 'sleep75' dataset**

```python
import numpy as np
import pandas as pd

from statsmodels.formula.api import ols

# Not to show Warnings
import warnings
warnings.simplefilter(action='ignore', category=Warning)
```

```python
# import new data
df = pd.read_csv('sleep75.csv')
```

```python
# specification sleep=b0+b1age+b2male 
# (how age and gender (male) are connected with sleep)
# y~ x1+x2+x3...
my_model = ols(formula = 'sleep~age+male', data = df).fit()
```

```python
my_model.params.round(3)
```

**REGRESSION**

sleep = 3146.745 + 3.589\*age-34.803\*male

### *Forecast the amount of sleep for a person who's 20 years old and male*

```python
# Plug in the Xs multiplying by the coeffs
3146.745 + 3.589*20-34.803*1
```

#### **Forecast**: according to our model a person who's 20 years old and male sleeps 3183.722  minutes a week

## **For the dataset wage1 consider a regression log(wage)  exper, exper2, female, married, smsa.**

```python
import numpy as np
import pandas as pd

from statsmodels.formula.api import ols

# Not to show Warnings
import warnings
warnings.simplefilter(action='ignore', category=Warning)
```

```python
df2 = pd.read_csv('wage1.csv')
```

```python
mod = ols(formula = 'np.log(wage)~exper+ I(exper**2)+ female + married + smsa', data = df2)
res = mod.fit()
```

```python
res.params.round(4)
```

#### Consider individuals with the following characteristics: 

|   |       |        |         |      |
|---|-------|--------|---------|------|
|   | exper | female | married | smsa |
| 1 | 5     | 1      | 1       | 1    |
| 2 | 26    | 0      | 0       | 1    |
| 3 | 38    | 1      | 1       | 0    |

#### Evaluate the prediction of the dependent variable for each individual.

**REGRESSION**

log(wage) =  1.250423 + 0.036880exper-0.000772exper^2
-0.362740female+0.143784married+0.272562smsa

```python
# compute log(wage) for person 1
# we need to plug in the values we have (year of experience, gender etc)
log_wage = 1.2504+0.0369*5-0.0008*np.square(5)-0.3627*1+0.1438*1+0.2726*1
log_wage.round(4)
```

```python
# 1.4686 is the natural log of the wage (remember we typed 'np.log(wage)' in the formula as Y), but we need the actual wage
# Thus we need the exponent
# wage = np.exp(1.4686)
np.exp(1.4686)
```

#### **Answer**: The predicted wage for person 1 (5 years of experience, female, married and lives in smsa) is 4.345

#### **The predicted wage for person 2**

```python
np.exp(1.2504+0.0369*26-0.0008*np.square(26)-0.3627*0+0.1438*0+0.2726*1)
```
